# Gravitational collapse

Gravitational collapse is a fascinating and complex phenomenon in astrophysics that occurs when a massive object, such as a star, reaches the end of its life cycle. It involves the force of gravity overpowering other forces, leading to the object's collapse under its own gravitational pull. 



1. **Equilibrium and Stability:**
Stars achieve equilibrium through a balance between the gravitational force pulling inwards and the pressure (thermal or radiation pressure) pushing outwards. This equilibrium is crucial for the star to maintain a stable size and shape throughout most of its life.

2. **Conditions for Gravitational Collapse:**
When a massive star exhausts its nuclear fuel, it can no longer produce the energy necessary to maintain the pressure that counteracts gravity. Depending on the mass of the star, various outcomes can occur, such as supernovae, neutron stars, or black holes.

3. **Hydrostatic Equilibrium Equation:**
The equilibrium of a star is governed by the hydrostatic equilibrium equation, which relates the pressure gradient to the mass density and gravitational force:

$$ \frac{dP}{dr} = - \frac{G M(r) \rho(r)}{r^2} $$

where

- $P$ = Pressure inside the star,
- $r$ = Radial distance from the center of the star
- $G$ = Gravitationa; constant
- $M(r)$ = Mass enclosed within the radius $r$
- $\rho(r)$ = Mass desnity at the radial distance

4. **Equation of State:**
The equation of state describes the relationship between the pressure, density, and temperature within the star. Different equations of state apply to different stages of a star's life, and they play a significant role in determining its fate during collapse.

5. **Chandrasekhar Limit:**
For stars supported by electron degeneracy pressure (white dwarfs), there is a maximum mass known as the Chandrasekhar limit. If a white dwarf exceeds this limit (approximately 1.4 times the mass of the Sun), it can no longer maintain electron degeneracy pressure and may undergo a type Ia supernova.

6. **Schwarzschild Radius:**
If the core of a massive star has a mass greater than the Tolman-Oppenheimer-Volkoff (TOV) limit, it will undergo a catastrophic gravitational collapse. The critical size at which a body becomes a black hole is given by the Schwarzschild radius:

$$R_s = \frac{2 G M}{c^2}$$

where 

- $R_s$ = Schwarschild radius
- $M$ = mass of the objects
- $c$ = speed of light in a vacuum.

7. **General Relativity:**
In the final stages of a massive star's collapse, General Relativity becomes essential to describe the extreme curvature of spacetime near the core, especially when a black hole forms.

8. **Numerical Simulations:**
Due to the complexity of gravitational collapse, numerical simulations using computational methods like hydrodynamics and numerical relativity are crucial for studying the detailed behavior of collapsing stars and the formation of black holes.

Studying gravitational collapse is a deeply intricate and mathematically demanding field. Astrophysicists use sophisticated mathematical models, equations, and simulations to gain a deeper understanding of this phenomenon and its consequences in the universe.

###  1D Monte Carlo simulation of particles undergoing gravitational collapse

- A  "1D" simulation means that we are considering a system in which the objects (particles in this case) can move only along a single straight line. 
- Each particle's position is determined by a single coordinate (e.g., the distance from a reference point), and the motion is constrained to that one-dimensional space.

- In the code example, we simulate gravitational collapse in a one-dimensional space, where particles are represented as points along a straight line. 
- This simplification allows us to focus on the core concepts of the Monte Carlo simulation and the gravitational interaction between particles.

In reality, astrophysical systems like stars and their collapses are three-dimensional, and a full understanding of gravitational collapse requires the consideration of three-dimensional space and the complex interactions between particles in 3D. However, implementing a full 3D simulation would be significantly more complex, and the one-dimensional example serves as a starting point to grasp the basic principles of Monte Carlo simulations in the context of gravitational collapse.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11    # Gravitational constant (m^3 kg^(-1) s^(-2))
M_sun = 1.989e30    # Mass of the Sun (kg)
R_sun = 6.957e8     # Radius of the Sun (m)

# Parameters
num_particles = 1000
mass_range = (0.5 * M_sun, 100 * M_sun)  # Mass range of particles
initial_density_range = (100, 500)       # Initial density range of particles (kg/m^3)
initial_velocity_range = (0, 100)        # Initial velocity range of particles (m/s)
total_time = 1.0e3                      # Total simulation time (s)
time_step = 1.0e2                      # Time step (s)

def initialize_particles():
    masses = np.random.uniform(mass_range[0], mass_range[1], num_particles)
    densities = np.random.uniform(initial_density_range[0], initial_density_range[1], num_particles)
    velocities = np.random.uniform(initial_velocity_range[0], initial_velocity_range[1], num_particles)
    return masses, densities, velocities

def calculate_gravitational_force(mass1, mass2, distance):
    return G * mass1 * mass2 / distance**2

def update_particle_positions(positions, velocities, time_step):
    return positions + velocities * time_step

def main():
    masses, densities, velocities = initialize_particles()
    positions = np.zeros(num_particles)
    time = 0

    while time < total_time:
        for i in range(num_particles):
            # Update positions
            positions[i] = update_particle_positions(positions[i], velocities[i], time_step)
            
            # Calculate net force on the particle due to gravitational interaction with others
            net_force = 0.0
            for j in range(num_particles):
                if i != j:
                    distance = np.abs(positions[i] - positions[j])
                    force = calculate_gravitational_force(masses[i], masses[j], distance)
                    net_force += force
            
            # Update velocities based on the net force
            acceleration = net_force / masses[i]
            velocities[i] += acceleration * time_step
            
        time += time_step

    # Plot the final positions of the particles
    plt.scatter(positions / R_sun, np.ones(num_particles), marker='o', s=5)
    plt.xlabel('Position (in Solar radii)')
    plt.title('Gravitational Collapse - Final Particle Positions')
    plt.show()

if __name__ == "__main__":
    main()
```

### 2D Monte Carlo simulation of gravitational collapse:
In this case, we will consider particles moving in a two-dimensional plane under the influence of gravitational forces. The rest of the code structure will remain similar, but we'll update the functions to handle two-dimensional positions and velocities.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11    # Gravitational constant (m^3 kg^(-1) s^(-2))
M_sun = 1.989e30    # Mass of the Sun (kg)
R_sun = 6.957e8     # Radius of the Sun (m)

# Parameters
num_particles = 1000  # Reducing the number of particles for lighter simulation
mass_range = (0.5 * M_sun, 100 * M_sun)
initial_density_range = (100, 500)
initial_velocity_range = (0, 100)
total_time = 1.0e4    # Reducing the total simulation time
time_step = 1.0e1     # Reducing the time step

def initialize_particles():
    masses = np.random.uniform(mass_range[0], mass_range[1], num_particles)
    densities = np.random.uniform(initial_density_range[0], initial_density_range[1], num_particles)
    velocities = np.random.uniform(initial_velocity_range[0], initial_velocity_range[1], size=(num_particles, 2))
    return masses, densities, velocities

def calculate_gravitational_force(mass1, mass2, distance):
    return G * mass1 * mass2 / distance**2

def update_particle_positions(positions, velocities, time_step):
    return positions + velocities * time_step

def main():
    masses, densities, velocities = initialize_particles()
    positions = np.zeros((num_particles, 2))
    time = 0

    while time < total_time:
        for i in range(num_particles):
            # Update positions
            positions[i] = update_particle_positions(positions[i], velocities[i], time_step)
            
            # Calculate net force on the particle due to gravitational interaction with others
            differences = positions - positions[i]
            distances = np.linalg.norm(differences, axis=1)
            distances[i] = 1.0  # Set distance to self to avoid division by zero
            forces = calculate_gravitational_force(masses[i], masses, distances)
            net_force = np.sum((forces / distances)[:, np.newaxis] * differences, axis=0)
            
            # Update velocities based on the net force
            acceleration = net_force / masses[i]
            velocities[i] += acceleration * time_step
            
        time += time_step

    # Plot the final positions of the particles
    plt.scatter(positions[:, 0] / R_sun, positions[:, 1] / R_sun, marker='o', s=5)
    plt.xlabel('X Position (in Solar radii)')
    plt.ylabel('Y Position (in Solar radii)')
    plt.title('Lighter Gravitational Collapse - Final Particle Positions (2D)')
    plt.show()

if __name__ == "__main__":
    main()
```

### Virial Theorem

The Virial Theorem is a fundamental result in classical mechanics that relates the average kinetic energy and the average potential energy of a system of particles in equilibrium. It provides insights into the stability and dynamics of physical systems, including astrophysical systems like stars undergoing gravitational collapse.

The Virial Theorem can be expressed in various forms, but one common form applicable to gravitational systems is:

$$\boxed{2\langle T \rangle + \langle V \rangle = 0}$$

where

- $\langle T \rangle$ = the avergae kinetc energy of the system
- $\langle V \rangle$ = the avergae potential energy of the system

In the context of gravitational collapse, let's consider a self-gravitating system, such as a star, in hydrostatic equilibrium. Hydrostatic equilibrium is a state in which the outward pressure force counterbalances the inward gravitational force, preventing further collapse.

**Application of the Virial Theorem to Gravitational Collapse:**

1. **Initial Equilibrium:** In the early stages of a star's life, it is in hydrostatic equilibrium, and the Virial Theorem holds. The star's gravitational potential energy is balanced by the thermal (kinetic) energy from nuclear fusion processes in the core.

2. **Nuclear Fusion Depletion:** As the star consumes its nuclear fuel and nuclear fusion processes slow down or cease, the production of thermal energy decreases.

3. **Gravitational Contraction:** With less thermal energy to counterbalance gravity, the star begins to contract under its self-gravitational force. As the core contracts, its temperature and pressure increase.

4. **Increased Gravitational Potential:** As the core contracts, the gravitational potential energy of the system increases.

5. **Violation of Virial Theorem:** In the collapsing phase, the average kinetic energy $\langle T \rangle$ decreases due to the depletion of thermal energy, while the gravitational potential energy ($\langle V \rangle$) increases. This leads to a violation of the Virial Theorem, and the relationship $2\langle T \rangle + \langle V \rangle = 0$ no longer holds.

6. **Stellar Collapse or Supernova:** If the star's core cannot generate enough thermal energy or other supporting forces to halt the collapse, it may undergo a gravitational collapse, leading to a supernova explosion or the formation of a compact object like a neutron star or black hole.

The violation of the Virial Theorem during gravitational collapse is a critical indicator that a system is no longer in equilibrium, and the balance between pressure and gravity is disrupted. This concept helps astronomers and physicists understand the conditions and stages leading to a star's collapse and its subsequent fate. Additionally, it plays a crucial role in theoretical astrophysics and the study of stellar evolution, providing valuable insights into the life cycles of stars.

## Derivation of Thermal Equilibrium Equation in Gravitational Collapse

In a spherically symmetric star undergoing gravitational collapse, we can derive an equation that represents the thermal equilibrium between pressure and gravity. This equation is an essential aspect of understanding how stars maintain stability and prevent further collapse.

- **Hydrostatic Equilibrium Equation**

    The hydrostatic equilibrium equation for a spherically symmetric star is given by:

    $$\frac{{dP}}{{dr}} = -\frac{{G M(r) \rho(r)}}{{r^2}}$$
    
    where:
    - $P$ is the pressure inside the star,
    - $r$ is the radial distance from the center of the star,
    - $G$ is the gravitational constant,
    - $M(r)$ is the mass enclosed within the radius $r$,
    - $\rho(r)$ is the mass density at the radial distance $r$.

- **Ideal Gas Equation of State**

    For an ideal gas, the pressure ($P$), density ($\rho$), and temperature ($T$) are related by:

    $P = n k T$

    where:
    - $n$ is the number density of particles (number of particles per unit volume),
    - $k$ is Boltzmann's constant.

    Expressing Number Density $n$ in terms of Mass Density $\rho$ and Mean Particle Mass $m$. The number density $n$ can be expressed in terms of the mass density $\rho$ and the mean particle mass $m$ as:

    $n = \frac{{\rho}}{{m}}$

    where 
    - $m$ is the mean particle mass (e.g., the proton mass in a fully ionized gas).

    Substituting the expression for $n$ into the ideal gas equation, we get: 
    $$P = \frac{{\rho k T}}{{m}}$$

- Next, we need to express the mass density $\rho$ in terms of the total mass ($M$) and the volume ($V$) of the star:
  
    $$\rho  = \frac{M}{V}$$

    The volume $V$ of a spherical shell of radius $r$ and thickness $dr$ is: $dV = 4 \pi r^2 dr$. Therefore, the volume of the star enclosed within radius $r$ is:

    $$V(r) = \int_0^r 4 \pi r'^2 dr' $$

    where the integration is performed from $r' = 0$ to $r$ (i.e., from the center of the star to the desired radius $r$).

  - Now, we can express the mass (M) enclosed within radius r in terms of the mass density () as:

    $$M(r) = \int_0^r 4\pi r'^2 \rho(r') dr'$$

    where again, the integration is from $r' = 0$ to r.

- Finally, we can substitute the expressions for $$ and $M(r)$ back into the hydrostatic equilibrium equation:

    $$\frac{d}{dr} \left(\frac{\rho k T}{m}\right) = - \frac{G}{r^2} \left(\int_0^r 4\pi r'^2 \rho(r') dr'\right) \frac{\rho}{V}$$

    After simplifying, we get:

    $$\frac{d}{dr} \left(\frac{k T}{m}\right) = - \frac{G}{r^2} \left(\int_0^r 4\pi r'^2  dr'\right)$$

    Now, let's integrate both sides of the equation from $r=0$ to $r$


    $$\int_0^r \frac{d}{dr} \left(\frac{kT}{m}\right) dr = - \int_0^r \frac{G}{r^2} \left(\int_0^r 4\pi r'^2 dr'\right) dr$$

    The left-hand side can be directly integrated:

    $$\frac{kT(r)}{m}-\frac{kT(0)}{m} = - \int_0^r \frac{G}{r^2} \left(\int_0^r 4 \pi r'^2 dr'\right) dr.$$

    SInce $T(0)$ is the temperature at the center of the star, we assume that it is constant (a central temprature for the star). Therefore, $kT(0)/m$ is just a constant. Let's denote it as $C$:

    $$C = \frac{kT(0)}{m}$$

    Now, we have:

    \begin{align*}
    \frac{{kT(r)}}{{m}} - C & = - \int_0^r \frac{G}{{r^2}} \left(\int_0^r 4 \pi r'^2 dr'\right) dr, \\
    & = C - \int_0^r \frac{G}{{r^2}} \left(\int_0^r 4 \pi r'^2 dr'\right) dr
    \end{align*}

    -  Now, we use the expression for the volume enclosed within radius $r$:

    $$V(r) = \int_0^r 4 \pi r'^2 dr'$$

    and rewrite the equation as:

    \begin{equation}
        \frac{kT(r)}{m} = C- \int_0^r \frac{G}{r^2} V(r') dr'
    \end{equation}

    This equation describe the thermal equilibrium in gravitational collapse. It shows that the temprature $(T(r))$ at a given radial distance ($r$) inside the star depends on the central temperature ($T(0)$), the gravitational constant, and the volume ($V(r')$) enclosed within each radial distance ($r'$).

    This equation illustrates that as the star contracts under its own gravity during gravitational collapse, the temperature at different radii changes to maintain thermal equilibrium. The pressure gradient (associated with temperature changes) must balance the gravitational forces to maintain hydrostatic equilibrium and prevent further collapse.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11      # Gravitational constant (m^3 kg^(-1) s^(-2))
k_boltzmann = 1.380649e-23  # Boltzmann's constant (m^2 kg s^(-2) K^(-1))
mean_particle_mass = 1.67e-27  # Mass of a proton (kg)

# Parameters
total_mass = 1.0 * 1.989e30    # Total mass of the star (kg)
star_radius = 1.0 * 6.957e8    # Radius of the star (m)
central_temperature = 10**7    # Central temperature (K)
num_points = 100               # Number of radial grid points

def calculate_volume(r):
    return 4 * np.pi * r**3 / 3

def calculate_temperature(r, constant):
    return constant - G * np.trapz(calculate_volume(np.linspace(0, r, num_points)) / r**2)

def main():
    # Calculate the constant using central temperature and ideal gas equation of state
    constant = k_boltzmann * central_temperature / mean_particle_mass

    # Calculate temperature profile at different radial distances
    radii = np.linspace(0, star_radius, num_points)
    temperatures = [calculate_temperature(r, constant) for r in radii]

    # Plot the temperature profile
    plt.plot(radii / 6.957e8, temperatures, label='Temperature (K)')
    plt.xlabel('Radius (in Solar radii)')
    plt.ylabel('Temperature (K)')
    plt.title('Temperature Profile in Thermal Equilibrium')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
